```python
import ast
ast.literal_eval('[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]')
```

```python
def convert(text):
    ext = []
    for i in ast.literal_eval(text):
        ext.append(i['name'])
    return ext    
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['genres']
```

```python
def convert1(text):
    ext1=[]
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 5:
            ext1.append(i['name'])
        counter +=1
    return ext1   
```

```python
movies['cast'] = movies['cast'].apply(convert1)
```

```python
def fetch_director(text):
    ext2 = []
    for i in ast.literal_eval(text):
        if i ['job']== 'Director':
            ext2.append(i['name'])
    return ext2        
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.dropna(inplace=True)
```

```python
def collapse(ext):
    ext4 = []
    for i in ext:
        ext4.append(i.replace(" ",""))
    return ext4    
```

```python
movies.head(2)
```

```python
movies['genres']= movies['genres'].apply(collapse)
movies['keywords']= movies['keywords'].apply(collapse)
movies['cast']= movies['cast'].apply(collapse)
movies['crew']= movies['crew'].apply(collapse)
```

# Feature Engineering/Extraction method - Bag of Words(BOW)

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(new['context']).toarray()
vector.shape
```

```python
pd.DataFrame(vector).head()
```

```python
cv.get_feature_names_out()
```

# Buliding Recommendation System basis Similarity Matrix (consin-similarity)

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
new
```

```python
len(new['title'])
```

```python
def recommendationsystem(movies):
    index = new[new['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key= lambda x: x[1])
    for i in distance[1:6]:
        print(new.iloc[i[0]].title)
```
